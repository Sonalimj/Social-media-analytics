```python
import requests
```

```python
from bs4 import BeautifulSoup #beautiful soup is used to parse the data into structure format,so we are using bs4 by importing beautiful soup
```

## header gives the user agent details regarding the information about search engine

```python
header={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36'}
```

```python
cookie={} ## we use cookies because python requests donot allow us to scrap the reviews until unless we set cookies and header.
```

## below function is used return the html page of the website which we want to scrap,if the status code is 200, it means we can access the code successfully. It helps to get the product name as well as the data ASIN number

```python
def getAmazonSearch(search_query):
    url="https://www.amazon.in/s?k="+search_query
    print(url)
    page=requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```

### below function also does the same thing to access the code successfully. It is used to get the customer review footer link

```python
def urlSearch(search_query):
    url="https://www.amazon.in/dp/"+search_query
    print(url)
    page=requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```

### search response shows the weblink page for hp laptop which is going to scrap.It shows all possible product of hp laptop

```python
search_response=getAmazonSearch("hp+laptops")
```

### below function is used to get the reviews and review rating

```python
def r(search_query):
    url="https://www.amazon.in"+search_query
    print(url)
    page=requests.get(url,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```

### below loop is used to extract the product name for 21 pages

```python
product=[]
for i in range(1,21):
    search_response=getAmazonSearch("hp+laptops&page="+str(i))
    soup=BeautifulSoup(search_response.content)
    for j in soup.findAll("span",attrs={'class':'a-size-medium a-color-base a-text-normal'}):
        product.append(j.text)
```

### below loop is used to scrap the data ASIN number for respective hp laptop for 21 pages

```python
tag=[]
for i in range(1,21):
    search_response=getAmazonSearch("hp+laptops&page="+str(i))
    soup=BeautifulSoup(search_response.content)
    for j in soup.findAll("div",attrs={'class':'sg-col-20-of-24 s-result-item sg-col-0-of-12 sg-col-28-of-32 sg-col-16-of-20 sg-col sg-col-32-of-36 sg-col-12-of-16 sg-col-24-of-28'}):
        tag.append(j['data-asin'])
        
```

### below loop is used to get the customer reviews link

```python
link=[]
for i in range(len(tag)):
    ##if tag[i]!="":
    res=urlSearch(tag[i])
    soup=BeautifulSoup(res.content)
    for j in soup.findAll('a',attrs={"data-hook":"see-all-reviews-link-foot"}):
        link.append(j['href'])
        
        
```

```python
len(link)
```

### below loop is used to get the reviews of customers from the above link function

```python
reviews=[]
for i in range(len(link)):
    response=r(link[i])
    soup=BeautifulSoup(response.content)
    for j in soup.findAll('span',attrs={"data-hook":"review-body"}):
        reviews.append(j.text)
```

### it gives the reviews along with customer ratings for 21 pages

```python
reviews=[]
ratings=[]
for i in range(len(link)):
    for k in range(1,21):
        response=r(link[i]+'&pageNumber='+str(k))
        soup=BeautifulSoup(response.content)
        for j in soup.findAll('span',attrs={"data-hook":"review-body"}):
            reviews.append(j.text)
        for star in soup.findAll('span',attrs={"data-hook":"review-star-rating"}):
            ratings.append(star.text)
```
